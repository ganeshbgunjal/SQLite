### SQLite Practice:

#### Importing sqlite:

```python
import sqlite3
```

#### Connection with db:

```python
connection = sqlite3.connect('College.db')
```

```python
connection
```

```python
cursor = connection.cursor()
```

#### Creating new table:

```python
cursor.execute('''
    create table Students (
        id integer primary key,
        fname text,
        lname text,
        class integer    
    )
''')

```

#### commit the changes:

```python
connection.commit()
```

#### insert data into table:

```python
cursor.execute('''
    insert into Students(id,fname,lname,class)
                values(1,'Ganesh','Gunjal',10),
                    (2,'Dilip','Gunjal',8),
                    (3,'Amol','Gunjal',5)
''')
```

```python
connection.commit()
```

```python
cursor.execute('select * from Students')
rows = cursor.fetchall()
for row in rows:
    print(row)
```
